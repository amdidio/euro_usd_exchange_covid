# EURO-USD Exchange Rate: COVID-19 Pandemic

For this project, I will be using a dataset which describes Euro daily exchange rates between 1999 and 2021, as recorded by the European Central Bank. For reference, if the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate). The Focus of this project will be on the exchange rate between the euro (€) and the American Dollar ($).

The full dataset can be found [here](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020).

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

The dataset contains 5,699 rows with 41 columns. Many columns contain varying amounts of null values, however, the `[US dollar ]` column does not seem to contain any null values. Finally, almost all columns have values stored as string objects.

## Data Cleaning

Before beginning my analysis, I will need to isolate the columns which I intend on using, as well as convert some of the values in these columns. Furthermore, I will also rename these particular columns for clarity.

```python
#renaming columns, converting to datetime, sorting by time
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
#isolating euro/USD exchange rate in new dataframe
euro_to_dollar = exchange_rates.copy()[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

While it initially seemed as though the US dollar did not contain any null values, a look at the value counts of this column shows 62 values of '-' in place of a numerical input. Since this is a very small portion of the dataset (about 1%), I will drop these moving forward.

```python
#dropping null values, converting to floats
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!='-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

Finally, I am left with 5,637 rows with 2 columns:
* `Time`: datetime object detailing when the exchange rate was recorded, from 01/04/1999 to 01/08/2021
* `US_dollar`: float object detailing the euro-dollar exchange rate.

## Visualizing the Evolution of the Euro-Dollar Exchange Rate

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.title('Euro-USD Exchange Rate Over Time')
plt.show()
```

To better visualize exchange rate trends for this time period, I will use a rolling means window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.suptitle('Euro-USD Exchange Rate Over Time')
plt.title('Rolling Window: 30 Days',size=10)
plt.show()
```

The exchange rate is at its lowest around 2001, and peaks around 2008.

## Visual Storytelling: Coronavirus Pandemic

From its first arrival, the coronavirus disease, or COVID-19, created a period of destablization worldwide. It's impact on the global economy left many countries reeling, and debating on how best to mitigate further losses. The United States was certainly no exception. 

To better explore the potential impact of COVID-19 on Euro-USD exchange rates, let's have a look at US case rates around the first year of the pandemic from [Our World in Data](https://ourworldindata.org/explorers/coronavirus-data-explorer?time=earliest..2021-01-08&facet=none&uniformYAxis=0&Metric=Confirmed+cases&Interval=7-day+rolling+average&Relative+to+Population=false&Color+by+test+positivity=true&country=~USA):



The above data begins January 28, 2020, shortly after the first recorded case of the Coronavirus disease in the US (1/20/2020), and ends January 8, 2021, the final date recorded in the dataset of Euro exchange rates. Looking at this data, there were 3 major waves of new COVID-19 cases, with periods of decreased cases in between:
* Wave 1: 3/17/2020 - 5/09/2020
* Wave 2: 6/12/2020 - 8/12/2020
* Wave 3: 10/12/2020 - End of Dataset (1/8/2021)

*Note: I will not be exploring potential causes for each of these waves of increased cases. However, more information on the COVID-19 timeline in the US from the CDC can be found [here](https://www.cdc.gov/museum/timeline/covid19.html).*

Now, let's have another look at the evolution of Euro-USD exchange rates during the same time period as above, and with the same rolling means window of 7 days:

```python
#isolating covid time period
covid_euro_dollar = (
    euro_to_dollar.copy()[euro_to_dollar['Time']
                   .between('2020-01-28','2021-01-08')]
)
covid_euro_dollar.reset_index(drop=True, inplace=True)
covid_euro_dollar['rolling_mean'] = covid_euro_dollar['US_dollar'].rolling(7).mean()


plt.plot(covid_euro_dollar['Time'],covid_euro_dollar['rolling_mean'])
#creating same xticks as above plot
plt.xticks([pd.to_datetime('2020-01-28'),pd.to_datetime('2020-04-30'),pd.to_datetime('2020-08-08'),pd.to_datetime('2021-01-08')],
           ['Jan 28, 2020','Apr 30, 2020','Aug 8, 2020','Jan 8, 2020']
          )
plt.title('Euro-USD Exchange Rate: Coronavirus Pandemic')
plt.show()
```

Looking at the two line plots, there does seem to be a correlation between new COVID-19 cases and the Euro-USD exchange rate during this time period. Let's better visualize this in order to get a sense of this correlation:

```python
import matplotlib.style as style
import matplotlib.dates as dates
from matplotlib.patches import ConnectionPatch

wave1 = covid_euro_dollar.copy()[covid_euro_dollar['Time']
                                 .between('2020-03-17','2020-05-09')]
wave2 = covid_euro_dollar.copy()[covid_euro_dollar['Time']
                                 .between('2020-06-12','2020-08-12')]
wave3 = covid_euro_dollar.copy()[covid_euro_dollar['Time']
                                 .between('2020-10-12','2021-01-08')]

style.use('seaborn')

### Adding the subplots ###
fig = plt.figure(figsize=(12, 7),facecolor='0.8')
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2,sharey=ax1)
ax3 = plt.subplot(2,3,3,sharey=ax1)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

### titles ###
fig.suptitle('EURO-USD rate continued to rise during COVID-19 pandemic, correlated with rising case numbers in the US'
             ,weight='bold')
fig.text(.081,.93,'EURO-USD exchange rates during phases of rising COVID-19 case rates in the first year of the pandemic (January 28, 2020 - January 8, 2021)')

### axes for all the subplots ###
for ax in axes:
    ax.set_ylim(1.075, 1.235)
    ax.set_yticks([1.075, 1.1, 1.125, 1.15,1.175,1.2,1.225])
    ax.set_yticklabels(['1.075', '1.100', '1.125', '1.150','1.175','1.200','1.225'],fontsize=8,alpha=0.3)

for ax in axes[:-1]:
    ax.set_xticks([])
    
### wave 1 plot ###
ax1.plot(wave1['Time'],wave1['rolling_mean'],
         color='c', lw=3)
ax1.set_title('WAVE 1',weight='bold',fontsize=10,color='c')
ax1.set_xlabel('March 17, 2020 - May 9, 2020',fontsize=8,alpha=.7)

### wave 2 plot ###
ax2.plot(wave2['Time'],wave2['rolling_mean'],
         color='m', lw=3)
ax2.set_title('WAVE 2',weight='bold',fontsize=10,color='m')
ax2.set_xlabel('June 12, 2020 - August 12, 2020',fontsize=8,alpha=.7)

### wave 3 plot ###
ax3.plot(wave3['Time'],wave3['rolling_mean'],
         color='y', lw=3)
ax3.set_title('WAVE 3',weight='bold',fontsize=10,color='y')
ax3.set_xlabel('October 12, 2020 - January 8, 2021',fontsize=8,alpha=.7)

### full plot ###
ax4.plot(covid_euro_dollar['Time'],covid_euro_dollar['rolling_mean'],lw=3)
ax4.set_xticklabels(['Mar 2020', 'May 2020','Jul 2020', 'Sep 2020','Nov 2020','Jan 2021'],
                    fontsize=8,alpha=.7)

### connection arrows ###
con1xy = (dates.date2num(pd.to_datetime('2020-03-17')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2020-03-17']['rolling_mean'].values[0])
con1 = ConnectionPatch(
    xyA=(con1xy[0]-2.3,1.076), coordsA=ax1.transData,
    xyB=con1xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con1)


con2xy = (dates.date2num(pd.to_datetime('2020-05-08')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2020-05-08']['rolling_mean'].values[0])
con2 = ConnectionPatch(
    xyA=(con2xy[0]+2.3,1.076), coordsA=ax1.transData,
    xyB=con2xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con2)


con3xy = (dates.date2num(pd.to_datetime('2020-06-12')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2020-06-12']['rolling_mean'].values[0])
con3 = ConnectionPatch(
    xyA=(con3xy[0]-2.3,1.076), coordsA=ax2.transData,
    xyB=con3xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con3)


con4xy = (dates.date2num(pd.to_datetime('2020-08-12')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2020-08-12']['rolling_mean'].values[0])
con4 = ConnectionPatch(
    xyA=(con4xy[0]+2.3,1.076), coordsA=ax2.transData,
    xyB=con4xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con4)


con5xy = (dates.date2num(pd.to_datetime('2020-10-12')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2020-10-12']['rolling_mean'].values[0])
con5 = ConnectionPatch(
    xyA=(con5xy[0]-4,1.076), coordsA=ax3.transData,
    xyB=con5xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con5)


con6xy = (dates.date2num(pd.to_datetime('2021-01-08')),
          covid_euro_dollar[covid_euro_dollar['Time']=='2021-01-08']['rolling_mean'].values[0])
con6 = ConnectionPatch(
    xyA=(con6xy[0]+4,1.076), coordsA=ax3.transData,
    xyB=con6xy, coordsB=ax4.transData,
    color='k',ls='--',lw=1,alpha=.5)
fig.add_artist(con6)

### color blocks ###
ax4.fill_between((pd.to_datetime('2020-03-17'),pd.to_datetime('2020-05-08')),
                 1.075,1.235,facecolor='c',alpha=0.2)
ax4.fill_between((pd.to_datetime('2020-06-12'),pd.to_datetime('2020-08-12')),
                 1.075,1.235,facecolor='m',alpha=0.2)
ax4.fill_between((pd.to_datetime('2020-10-12'),pd.to_datetime('2021-01-08')),
                 1.075,1.235,facecolor='y',alpha=0.2)

### signature ###
fig.text(.08, .04,'©Alexa DiDio' + ' '*91 + 'Source: European Central Bank, Our World in Data',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=12)

plt.show()
```

While the second and third waves of rising COVID-19 cases seem to greatly correlate with EURO-USD exchange rate, the same cannot be said of the first wave of cases from this visual. However, we're really only seeing part of the picture here. In order to understand what was happening during that first wave, we need to look at case numbers in European countries (specifically those that use the Euro as currency). 



Above is a snapshot of European case rates between March and July of 2020, alongside US case rates for comparison (again, from [Our World in Data](https://ourworldindata.org/explorers/coronavirus-data-explorer?time=2020-03-01..2020-07-01&facet=none&uniformYAxis=0&Metric=Confirmed+cases&Interval=7-day+rolling+average&Relative+to+Population=true&Color+by+test+positivity=true&country=USA~FRA~ESP~DEU~NLD~ITA)). For this sample, I chose the five wealthiest European countries in 2020 (with the Euro as currency) by GDP ([source](https://en.wikipedia.org/wiki/List_of_sovereign_states_in_Europe_by_GDP_(nominal))).

The first wave in rising cases for the US was identified around March 17 - May 9, but we can see that this same time period was equally (if not more) chaotic for Europe. Therefore, it's not hard to imagine why the Euro-USD exchange rate might remain roughly stagnant during this period. However, by the time we enter June 2020, the case rates for these European countries have more or less bottomed out. The US case rates, on the other hand, only plateaued until its second wave of rising case rates began around June 12. Thus, it is also not surprising that the second wave US case rates has a much stronger correlation with the Euro-USD exchange rate.

Overall, the Euro-USD exchange rate continued to rise in a linear fashion, alongside a roughly exponential rise in COVID-19 cases in the US, suggesting a strong correlation between the two.
